# Missingno - Missing Data

```python
import missingno as msno
msno.matrix(df)
plt.show()
```

# Seaborn - Heatmap
### Relationship between columns
* 01 --> Direct proportion
* 00 --> No relationship
* -1 --> İnverse proportion

```python
f,ax = plt.subplots(figsize=(15, 10))
sns.heatmap(df.corr(), annot=True, linewidths=0.5,linecolor="red", fmt= '.1f',ax=ax)
plt.show()
```

# Separating Features and Labels
### X Feature = SepalLengthCm ,SepalWidthCm ,PetalLengthCm ,PetalWidthCm

```python
X=df.iloc[:, 1:5]
X.head()
```

# Converting String Value To int Type for Labels
### Encode label category
* Iris-setosa -> 0
* Iris-versicolor -> 1
* Iris-virginica -> 2

```python
df.Species.unique()
```

```python
from sklearn.preprocessing import LabelEncoder
y=df.iloc[:,-1]

encoder = LabelEncoder()
y = encoder.fit_transform(y)
print(y)
```

# Data Standardisation

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

# Splitting Dataset into Training Set and Testing Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

# Build KNN Model with Default Hyperparameter

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

classifier = KNeighborsClassifier(n_neighbors = 10)
classifier.fit(X_train, y_train)
y_pred=classifier.predict(X_test)
```

# Accuracy Score

```python
print('Accuracy Score:')
print(metrics.accuracy_score(y_test,y_pred))
```

### Highest Accuracy Score

# Confusion Matrix with Seaborn - Heatmap

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
f,ax = plt.subplots(figsize=(10, 10))
sns.heatmap(cm, annot=True, linewidths=0.5,linecolor="red", fmt= '.0f',ax=ax)
plt.show()
```
